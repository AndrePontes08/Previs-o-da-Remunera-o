## Etapa 3 Crisp-DM: Preparação dos dados
Nessa etapa realizamos tipicamente as seguintes operações com os dados:

 - **seleção**: Já temos os dados selecionados adequadamente?
 - **limpeza**: Precisaremos identificar e tratar dados faltantes
 - **construção**: construção de novas variáveis
 - **integração**: Temos apenas uma fonte de dados, não é necessário integração
 - **formatação**: Os dados já se encontram em formatos úteis?


```python
df.drop(['Unnamed: 0', 'id_cliente'],inplace=True, axis=1)
```

```python
df.isna().sum()
```

```python
#porcentagem de NaN
df['tempo_emprego'].isna().sum() / len(df)
```

```python
#preenche os NaN com a media da variavel tempo_emprego
media = df['tempo_emprego'].mean()
df['tempo_emprego'].fillna(media, inplace=True)

df.isna().sum()
```

```python
metadata = pd.DataFrame(df.dtypes, columns = ['tipo'])

metadata['n_categorias'] = 0

for var in metadata.index:
    metadata.loc[var,'n_categorias'] = len(df.groupby([var]).size())
    
metadata
```

```python
#transforma as variaveis categoricas em dummys
#.columns.tolist(): Converte os nomes dessas colunas em uma lista

colunas_categoricas = df.select_dtypes(include=['object', 'category']).columns.tolist()
renda_d = pd.get_dummies(df, columns=colunas_categoricas, drop_first=True)
```

## Etapa 4 Crisp-DM: Modelagem
Nessa etapa que realizaremos a construção do modelo. Os passos típicos são:
- Selecionar a técnica de modelagem
- Desenho do teste
- Avaliação do modelo


### Rodando o modelo

```python
renda_d.shape
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
from sklearn.model_selection import GridSearchCV
```

```python
X = renda_d.drop('renda', axis=1)
y = renda_d['renda']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1997)
```

### Buscando os  melhores parâmetros

```python
param_grid = {
    'n_estimators': [100, 300, 500],
    'max_depth': [10, 20, 30],
    'min_samples_split': [2, 5, 10],
    'max_features': ['auto', 'sqrt', 'log2']
}

grid_search = GridSearchCV(estimator=RandomForestRegressor(random_state=1997), param_grid=param_grid, cv=5, n_jobs=-1, verbose=2)
grid_search.fit(X_train, y_train)

print(grid_search.best_params_)
```

```python
regressao = RandomForestRegressor(min_samples_split=5,n_estimators=100,max_features='auto',max_depth=20,random_state=1997)
regressao.fit(X_train, y_train)
```

```python
y_pred = regressao.predict(X_test)
y_train_pred = regressao.predict(X_train)
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
r2_train = r2_score(y_train, y_train_pred)
print(f'MSE: {mse}')
print(f'R2 score teste: {r2}')
print(f'R2 score treino: {r2_train}')
```

## Etapa 5 Crisp-DM: Avaliação dos resultados


Com base no R2 o modelo está prevendo corretamente 47,3% na base de teste
e 76,6% na base de treino
